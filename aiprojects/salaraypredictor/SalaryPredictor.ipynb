```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv("salaries_dataset.csv", sep = ";")
```

```python
df
```

```python
plt.scatter(df['experience_level'], df['salary'])
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(df[['experience_level']], df['salary'])

plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.scatter(df['experience_level'], df['salary'])

x = df['experience_level']
y = reg.predict(df[['experience_level']])
plt.plot(x, y, color = "green", label = "Linear Regression")
plt.legend()
plt.show()
```

```python
# Call PolynomialFeatures function to create a polynomial regression object
# Degree is specified when calling this function
polyreg = PolynomialFeatures(degree = 5)

xpoly = polyreg.fit_transform(df[['experience_level']])
```

```python
reg = LinearRegression() 
reg.fit(xpoly, df['salary'])
```

```python
yhead = reg.predict(xpoly)
plt.plot(df['experience_level'], yhead, color = 'red', label = "polynomial regression")
plt.legend()

plt.scatter(df['experience_level'], df['salary'])
plt.show()
```

```python
xpoly1 = polyreg.fit_transform([[4.5]])
reg.predict(xpoly1)
```
